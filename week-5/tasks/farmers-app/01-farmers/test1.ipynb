# 😥 problem
- easy to get id of a product: `product.id`
- but no way to know what product is for `id=4`

# 🌟 solution
- data is stored in tables
- tables has rows and cols
- each row has `#` col, and a corresponding `row object`
- `row object` will have attributes for cols
- such a table in python ~ dict with keys as slot

👉 to implement such solution, we've to plan from scratch: [test2.ipynb](./test2.ipynb)

```python
class Product:
  _idx = 1

  def __init__(self, name, type):
    self.id = Product._idx
    Product._idx += 1

    self.name = name
    self.type = type  # dairy, vegetable, fruit, nut, others

  def __repr__(self):
    return f'<Product {self.name}>'


potato = Product('potato', type='vegetable')
tomato = Product('tomato', type='vegetable')
milk = Product('milk', type='dairy')
honey = Product('honey', type='others')

honey
honey.id
```

```python
class Stock:
  _idx = 1

  def __init__(self, product_id, qty, price, expiry):
    self.id = Stock._idx
    Stock._idx += 1

    self.product_id = product_id
    self.qty = qty
    self.price = price
    self.expiry = expiry

  def __repr__(self):
    return f'<Stock {self.product_id}>'


potato_stock1 = Stock(product_id=1, qty=50, price=40, expiry=20)
potato_stock2 = Stock(product_id=1, qty=150, price=20, expiry=5)
honey_stock = Stock(product_id=4, qty=10, price=500, expiry=500)

honey_stock
```

```python
inventory = {
    1: potato_stock1,
    2: potato_stock2
}


def add_stock(stock):
  inventory[stock.id] = stock


add_stock(honey_stock)
inventory
```
