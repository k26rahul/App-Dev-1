```python
def decode(encoding_scheme, size, file='./raw-binary.txt'):
  raw_binary = open(file).read().strip()
  bit_groups = [raw_binary[i:i+size] for i in range(0, len(raw_binary), size)]
  message = ''
  for bit_group in bit_groups:
    if len(bit_group) < size:
      continue
    letter = encoding_scheme[bit_group]
    message += letter
  return message
```
